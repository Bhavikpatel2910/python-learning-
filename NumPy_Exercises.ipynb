**Student Marks Analysis (Data Processing)**

```python
import numpy as np

marks = np.array([
    [85, 78, 92, 88],
    [76, 81, 79, 85],
    [90, 92, 94, 91],
    [65, 70, 72, 68],
    [88, 84, 91, 87]
])
marks
```

```python
# Find average marks per student
averagemarks = marks.mean(axis=1)
averagemarks
```

```python
# Find highest scorer

highestscorer = marks.max(axis=1)
highestscorer
```

```python
# Find subject-wise average
averagemarks = marks.mean(axis=0)
averagemarks
```

```python
# Add 5 bonus marks to all students
bonusmarks = 5
marks += bonusmarks
marks
```

```python
# Find students scoring >85 overall

a = marks[:4,3]
for i in a:
  if i > 85:
    print(i)
```

**E-commerce Sales Data (Business Analytics)**

```python
sales = np.array([200, 220, 250, 300, 280, 260, 310])
```

```python
# Calculate total revenue

total=sales.sum()
total
```

```python
# Find day with maximum sales

maxsales=sales.max()
maxsales
```

```python
# Compute moving average (window=3)
window = 3

c = np.cumsum(sales, dtype=float)

moving_avg = (c[window:] - c[:-window]) / window

moving_avg
```

```python
# Normalize sales data (0–1 scale)
normalized_sales = (sales - sales.min()) / (sales.max() - sales.min())
normalized_sales
```

```python
# Find percentage increase each day
percentage= sales[:6]
p =0
for i in percentage:
  a =((i-p)/p)*100
  p = i
  print(a)

```
