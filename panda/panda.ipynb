**Series in Panda**

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a','b','c']
list = [10,20,30]
arr = np.array([10,20,30])
dic = {1:20,2:20,3:30}
```

```python
pd.Series(list)
```

```python
pd.Series(list,index=labels)
```

```python
pd.Series(arr)
```

```python
pd.Series(arr,index=list)
```

```python
pd.Series(dic,index=labels)
```

```python
pd.Series(dic,index=labels)
```
