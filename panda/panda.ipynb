**Series in Panda**

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a','b','c']
list = [10,20,30]
arr = np.array([10,20,30])
dic = {1:20,2:20,3:30}
```

```python
pd.Series(list)
```

```python
pd.Series(list,index=labels)
```

```python
pd.Series(arr)
```

```python
pd.Series(arr,index=list)
```

```python
pd.Series(dic)
```

```python
data = {
    'name' : ['bhavik','savan','parth','lav'],
    'age' :[19,18,17,19],
    'city' : ['ahemdabad','gandhinagr','maninagar','anand'],
    'salary':[10000,5000,2000,4000]
}
df = pd.DataFrame(data)
```
